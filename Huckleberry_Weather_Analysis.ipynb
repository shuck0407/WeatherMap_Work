```python
#This is the main processing cell.  It retrieves the weather data for each city in our city array 

# Create an empty list to hold the data for each city in our cities list
weather_data = [] 


#First print a message saying we're starting the API calls

print(f"Beginning Openweathermap Data Retrieval \n")

#Count the number of calls we make to the API.  We need to initialize a counter for this.
#We also want to keep track of the number of errors we get so create a counter for that too.
city_counter = 1
error_counter = 0

for index, row in random_cities_df.iterrows():
    city = row['city']
    
    print(f'Processing Record {city_counter} for city {city}.')
    print(f'{ow.BASE_URL}weather?q={city}&units=imperial&APPID={api_key}')
    
    try:
        response = ow.get_current(city, **settings) 
        if response("cod") != 200:
            print('Return code not equal to 200.  Skipping..... \n')
            error_counter= error_counter + 1
            
        else:
            weather_data.append(response)
        
    except:
        print('City not found.  404 error.  Skipping..... \n')
        error_counter= error_counter + 1
           
            
        
    #Increment the city counter    
    city_counter = city_counter + 1
    
print('----------------------------- \n')
print('Data Retrieval Complete \n') 
print('----------------------------- \n')
print(f'Could not get data for {error_counter} cities')

```

```python
#Load all of the weather data into a dataframe

city_weather_df = pd.DataFrame(weather_data)
city_weather_df = city_weather_df.drop(columns=['sys', 'cod', 'base', 'weather', 'rain'])

#reorganize the columns
city_weather_df = city_weather_df[['name', 'id', 'coord', 'dt', 'main', 'clouds', 'wind', 'visibility']]
city_weather_df

#merge the country data into the city weather dataframe in case we need country in our analysis
weather_df = city_weather_df.rename(columns={'name': 'city'})
weather_df = pd.merge(weather_df, random_cities_df, on='id')

#drop the duplicate city column and rename city_x

weather_df = weather_df.drop(columns=['city_y'])
weather_df = weather_df.rename(columns={'city_x':'city'})
#reorganize the columns again
weather_df = weather_df[['city', 'country', 'id', 'coord', 'dt', 'main', 'clouds', 'wind', 'visibility']]

weather_df
```

```python
#reorganize dataframe to remove dictionaries from columns and make new columns

lat_lon_df = weather_df.join(pd.DataFrame(weather_df.pop('coord').tolist()))
lat_lon_df

wind_df = lat_lon_df.join(pd.DataFrame(lat_lon_df.pop('wind').tolist()))
wind_df

cloud_df = wind_df.join(pd.DataFrame(wind_df.pop('clouds').tolist()))
cloud_df

main_weather_df = cloud_df.join(pd.DataFrame(cloud_df.pop('main').tolist()))
main_weather_df

#drop some more unnecessary columns
main_weather_df = main_weather_df.drop(columns=['grnd_level', 'sea_level', 'temp', 'temp_min'])


#rename the columns
main_weather_df = main_weather_df.rename(columns={'deg': 'wind_direction', 'speed': 'wind_speed', 'all' : 'cloudiness'})
main_weather_df
```

```python
#Send the main weather dataframe to a csv file

main_weather_df.to_csv("Huckleberry_Weather.csv", encoding='utf-8', index=False)
```

```python
#Get the UTC date from our dataframe for use in our plot titles later

#Just grab the first date in the dataframe since every row was retrived around the same time
current_dt = (main_weather_df.loc[0][3])

date_title = datetime.datetime.fromtimestamp(current_dt).strftime('%m-%d-%y %H:%M:%S')
date_title
```

## Latitude vs. Temperature Plot

```python
#set the overall plot size
plt.figure(figsize=(14,10))

sea.set_style('darkgrid')

#Create a scatter plot that compares latitude versus the maximum current temperature
# the zorder argument puts the dots in back of the grid so that the grid shows up

sea.scatterplot(x='lat', y='temp_max', data=main_weather_df, zorder=2, edgecolors='w', s=200)
plt.axvline(x=0, label='Equator', linestyle='--', color='g')
plt.xlabel('Latitude')
plt.ylabel('Max Temperature (F)')
plt.title(f'City Latitude vs Max Temperature ({date_title})', fontsize=24)
plt.text(0, 70, 'Equator', rotation=90, verticalalignment='center', fontsize=16)
plt.grid(zorder=3)
#Save the plot and show it
plt.savefig("Latitude_Vs_Temp.png")
plt.show()
```
